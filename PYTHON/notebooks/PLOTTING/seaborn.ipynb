Gráficos de Matriz

```python
# Seleccionando solo las columnas numéricas
numerical_columns = df.select_dtypes(include=['number'])

# Calculando la matriz de correlación para las columnas numéricas
correlation_matrix = numerical_columns.corr()

sns.heatmap(data=correlation_matrix, cmap="YlGnBu", vmin = -1, vmax = 1)
```

FacetGrid

```python
g = sns.FacetGrid(df, col="Department", row="Education")
g.map(sns.histplot, "JobSatisfaction")
```

```python
# Creando un objeto FacetGrid, definiendo las columnas y filas según el departamento y el nivel de educación
g = sns.FacetGrid(df, col="Department", row="Education", margin_titles=True)

# Aplicando un histograma de la satisfacción laboral en cada faceta
g.map(sns.histplot, "JobSatisfaction", bins=10, color="skyblue", edgecolor="black")

# Añadiendo títulos y etiquetas
g.set_axis_labels("Satisfacción Laboral", "Frecuencia")
g.set_titles(col_template="{col_name} Departamento", row_template="Educación {row_name}")
g.set(xlim=(0, df['JobSatisfaction'].max()), ylim=(0, None))

# Ajustando el diseño
g.tight_layout()

# Mostrando el gráfico
plt.show()
```

Relación con Matplotlib

Solapamiento: Seaborn está construido sobre Matplotlib, por lo que puedes utilizar funciones y métodos de Matplotlib junto con Seaborn para personalizar aún más los gráficos.

Complementariedad: Seaborn simplifica la creación de muchos tipos de gráficos que requieren múltiples líneas de código en Matplotlib. Ofrece una interfaz más sencilla y estilizada.

Diferencias: Aunque Seaborn facilita la creación de gráficos estadísticos complejos, Matplotlib ofrece más control sobre los detalles de bajo nivel de los gráficos.

Estilos y Paletas: Seaborn viene con estilos y paletas de colores predefinidos que hacen que los gráficos sean visualmente atractivos de manera predeterminada.

En resumen, Seaborn amplía las capacidades de Matplotlib, ofreciendo una interfaz más amigable y estilizada para crear gráficos estadísticos. Puedes utilizar ambas bibliotecas juntas para aprovechar la facilidad de uso de Seaborn y el control detallado de Matplotlib.

# Estilos y colores

Seaborn facilita la creación de gráficos atractivos y coherentes a través de la utilización de estilos y paletas de colores. Veamos cómo puedes utilizar estos elementos en Seaborn:

1. Estilos Predefinidos

Seaborn ofrece una serie de estilos predefinidos que puedes aplicar a tus gráficos. Estos estilos controlan la apariencia de los elementos del gráfico, como el fondo, las líneas de cuadrícula y los colores.

Puedes aplicar un estilo utilizando la función sns.set_style. Aquí hay algunos estilos comunes:

    + "white": Fondo blanco sin líneas de cuadrícula.
    + "dark": Fondo oscuro sin líneas de cuadrícula.
    + "whitegrid": Fondo blanco con líneas de cuadrícula.
    + "darkgrid": Fondo oscuro con líneas de cuadrícula.
    + "ticks": Fondo blanco con marcas en los ejes.

```python
sns.set_style("whitegrid")
sns.boxplot(x="Department", y="Age", data=df)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.boxplot(x="Department", y="Age", data=df)
plt.show()
```

Paletas de Colores

Seaborn también ofrece una variedad de paletas de colores que puedes utilizar para colorear los elementos de tu gráfico. Las paletas de colores son conjuntos coherentes de colores que funcionan bien juntos.

Puedes aplicar una paleta de colores utilizando la función sns.set_palette. Algunas paletas comunes incluyen:

    + "deep"
    + "muted"
    + "bright"
    + "pastel"
    + "dark"
    + "colorblind"

```python
sns.set_palette("pastel")
sns.barplot(x="Department", y="Age", data=df)
plt.show()
```

```python
sns.set_palette("deep")
sns.barplot(x="Department", y="Age", data=df)
plt.show()
```

Las paletas de colores en Seaborn están diseñadas para diferentes propósitos y contextos, y elegir la paleta adecuada puede ayudar a transmitir mejor la información en tu gráfico. A continuación, te describo algunas paletas comunes y sus usos típicos:

1. "deep"

Apropiada para: Gráficos generales con colores distintivos y saturados.

Uso: Cuando necesitas diferenciar claramente entre varias categorías.

2. "muted"

Apropiada para: Gráficos que requieren colores menos llamativos.

Uso: Similar a "deep", pero con tonos más suaves, lo que puede ser útil cuando se desea un aspecto más sutil.

3. "bright"

Apropiada para: Gráficos que necesitan colores vivos y brillantes.

Uso: Cuando se desea destacar categorías o elementos específicos en los datos.

4. "pastel"

Apropiada para: Gráficos que requieren colores suaves y agradables.

Uso: Ideal para presentaciones donde los colores brillantes y saturados pueden ser demasiado abrumadores.

5. "dark"

Apropiada para: Gráficos que requieren colores oscuros y saturados.

Uso: Puede ser útil para destacar elementos en un fondo claro.

6. "colorblind"

Apropiada para: Gráficos destinados a una audiencia que pueda incluir personas con daltonismo.

Uso: Asegura que los colores sean distinguibles para aquellos con deficiencias de visión de color.

7. Paletas Secuenciales

Apropiada para: Representar datos ordenados, donde hay una progresión clara de un valor bajo a uno alto.

Uso: Mapas de calor, gráficos de contorno y cualquier visualización que represente una secuencia o gradiente.

8. Paletas Divergentes

Apropiada para: Datos donde hay un punto medio interesante y valores divergentes a ambos lados.

Uso: Comparación de desviaciones respecto a un punto central, como en gráficos de diferencias o anomalías.

9. Paletas Cúbicas Helix (cubehelix)

Apropiada para: Representar datos en escala de grises con un cambio en brillo.

Uso: Gráficos que serán impresos en blanco y negro o para aquellos con deficiencias de visión de color.

Combinación de Estilos y Paletas

Puedes combinar estilos y paletas de colores para crear una apariencia coherente en todos tus gráficos. La función sns.set te permite configurar ambos a la vez:

```python
sns.set(style="darkgrid", palette="muted")
sns.barplot(x="Department", y="Age", data=df)
plt.show()
```

Contexto de Dibujo

Seaborn también te permite ajustar el contexto de dibujo, lo cual afecta el tamaño y la complejidad de los elementos del gráfico (por ejemplo, el tamaño de las etiquetas). Puedes utilizar sns.set_context para esto, con opciones como "paper", "notebook", "talk", y "poster".

Resumen

Los estilos y paletas de colores en Seaborn te permiten crear gráficos atractivos y coherentes con muy poco esfuerzo. Al aplicar estos elementos, puedes garantizar que tus gráficos sean fáciles de leer y visualmente agradables, sin tener que ajustar manualmente muchos detalles.
